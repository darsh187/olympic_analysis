```python
import pandas as pd
import numpy as np
```

```python
ath_df = pd.read_csv('athlete_events.csv')
reg_df = pd.read_csv('noc_regions.csv')
```

```python
ath_df = ath_df[ath_df['Season'] == 'Summer']
```

```python
ath_df.shape
```

```python
reg_df.tail()
```

```python
df = ath_df.merge(reg_df, on='NOC', how='left')
```

```python
df.tail()
```

```python
df['region'].nunique()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df = pd.concat([df,pd.get_dummies(df['Medal'])], axis=1)
```

```python
df.tail()
```

```python
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index().head(20)
```

```python
medal_telly = df.drop_duplicates(subset=['Team','NOC','Games','Year','Season','City','Sport','Event','Medal'])
```

```python
medal_telly = medal_telly.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```
